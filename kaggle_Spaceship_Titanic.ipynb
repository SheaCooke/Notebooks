```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.impute import SimpleImputer

test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
```

```python
print(train.columns)

print(train.isnull().values.any())
```

```python
train.describe(include='all')
```

```python
train.head()
```

```python
#drop name from training set, so also drop it from test set
#cabin is almost always unique -> no predictive power, should be dropped along with passengerId and Name

train_cleaned = train.drop(['Cabin', 'Name'], axis=1)
test_cleaned = test.drop(['Cabin', 'Name'], axis=1)

```

```python
# encode true / false columns: VIP, CryoSleep
train_cleaned['VIP'] = train_cleaned['VIP'].map({'True': 1, 'False': 0})
train_cleaned['CryoSleep'] = train_cleaned['CryoSleep'].map({'True': 1, 'False': 0})

test_cleaned['VIP'] = test_cleaned['VIP'].map({'True': 1, 'False': 0})
test_cleaned['CryoSleep'] = test_cleaned['CryoSleep'].map({'True': 1, 'False': 0})
```

```python
# encode categorical features HomePlanet, Destination
train_cleaned = pd.get_dummies(train_cleaned, columns=['HomePlanet', 'Destination'])
```
